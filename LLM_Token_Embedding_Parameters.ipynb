```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
#training data
sentences=[
"the flutty cat that was sleeping under the table woke up and scretched",
"the old dog that had been chasing squirrels all day finally rested",
"the small bird that flew away from the nest restuned to its mother",
"the big house that stood on hill overlooked the entire vallay",
"the curious child who found a hidden door opened it carefully",
"the talented musician who played the vilion performed a beautiful piece",
"the adventurous traveler who explored the ancient ruins discovered a secret passage",
"the dedicated scientist who studied the rare flower found a new species"
]
```

```python
# create vocabulary
words=set()
for sentence in  sentences:
   for word in sentence.split():
       words.add(word)
words=list(words)
```

```python
words
```

```python
word_to_index={word:i for i ,word in enumerate(words)}
index_to_word={i:word for i , word in enumerate(words)}
```

```python
word_to_index # word side index number display
```

```python
# index number is display on first the side ways word display
index_to_word
```

```python
#create eords in PAIRS of TWO TWO

#store vocab
vocab_size=len(words)
```

```python
vocab_size # the 72 is the uqnic words in the Training data
```

```python
# create list
x_pairs = []
y_next =  []
```

```python
  # create Pairs
for sentence in sentences:
    sentence_words =sentence.split()
    for i in range(len(sentence_words)-2):
      x_pairs.append([word_to_index[sentence_words[i]] ,word_to_index[sentence_words[i + 1]]])
      y_next.append(word_to_index[sentence_words[i+2]])
```

```python
x_pairs # this out linked to the word_to_index numbers
```

```python
y_next
```

```python
# create test data
x_pairs = np.array(x_pairs)
y_next = np.array(y_next)
```

```python
# split the data
x_train,x_test,y_train,y_test=train_test_split(x_pairs,y_next,test_size=0.2,random_state=42)
```

```python
# embeding dimension size
embedding_dim=5
```

```python
# instalize keras model
model=tf.keras.Sequential([
   # add layers & embedding
   tf.keras.layers.Embedding(vocab_size,embedding_dim), #vocab-size=72,emabeding_dim=5
   # add flatten layers
   tf.keras.layers.Flatten(),
   #add layers
   tf.keras.layers.Dense(8,activation='relu'),
   tf.keras.layers.Dense(vocab_size,activation='softmax')

])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=20,verbose=1)
```

```python
model.summary()
```

```python
input_pair=np.array([[word_to_index["the"],word_to_index["cat"]]]) #(1,2)
predictions=model.predict(input_pair)
predictions
```

```python
predicted_index=np.argmax(predictions[0])
print(f"RNN: Predicted next word for 'the cat': {index_to_word[predicted_index]}")
```
